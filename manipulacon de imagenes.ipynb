# MANIPULACION DE IMAGENES


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
imagen = cv2.imread('cat_or_dog_4.jpg')
plt.imshow(imagen)
```

```python
# cambio de orden de RGB
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
plt.imshow(imagen)
```

```python
imagen.shape
```

## cambiar tamao de la imagen

```python
imagen_2 = cv2.resize(imagen, (600, 500))
plt.imshow(imagen_2)
print(imagen_2.shape)
```

###  cambio de tamao por Porcentaje

```python
radio_ancho = 0.5
radio_alto = 0.5
```

```python
imagen_2 = cv2.resize(imagen_2, (0,0),imagen_2,radio_ancho, radio_alto)
plt.imshow(imagen_2)
print(imagen_2.shape)
```

### girar imagenes

```python
# CAMBIAR DE DERECHA A IZQUIERDA
imagen_2 = cv2.flip(imagen_2,1)
plt.imshow(imagen_2)
print(imagen_2.shape)
```

```python
# CAMBIAR DE ARRIBA  A ABAJO
imagen_2 = cv2.flip(imagen_2, 0)
plt.imshow(imagen_2)
print(imagen_2.shape)
```

```python
# GUADAR IMAGEN MODIFICADA 
imagen_2 = cv2.cvtColor(imagen_2, cv2.COLOR_BGR2RGB)
```

```python
cv2.imwrite('gato_alrrebes.jpg', imagen_2)
```
